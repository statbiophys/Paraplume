```python
import pandas as pd
from collections import defaultdict
def rec_dd():
    return defaultdict(rec_dd)
import json
```

```python
paragraph_labels = pd.read_csv("/home/gathenes/paragraph_benchmark/paragraph_labels/processed_dataset_paragraph.csv")
```

```python
print(paragraph_labels.columns)
```

```python
paragraph_label_dict = rec_dd()
for pdb, group in paragraph_labels.groupby("pdb"):
    pdb=pdb.lower()
    for chain, group_chain in group.groupby("chain_type"):
        chain_label=["light labels","heavy labels"][chain=="H"]
        chain_sequence=["light sequence","heavy sequence"][chain=="H"]
        paragraph_label_dict[pdb][chain_label]=group_chain["paratope"].values[0]
        paragraph_label_dict[pdb][chain_sequence]=group_chain["sequence"].values[0]
    paragraph_label_dict[pdb]=dict(paragraph_label_dict[pdb])
paragraph_label_dict=dict(paragraph_label_dict)
```

```python
test="/home/gathenes/paragraph_benchmark/241107/test_set/dict.json"
train="/home/gathenes/paragraph_benchmark/241107/train_set/dict.json"
val="/home/gathenes/paragraph_benchmark/241107/val_set/dict.json"
with open(test) as f:
    test_dict=json.load(f)
with open(train) as f:
    train_dict=json.load(f)
with open(val) as f:
    val_dict=json.load(f)
```

```python
my_label_dict=rec_dd()
for type in [train_dict,test_dict, val_dict]:
    for values in type.values():
        pdb=values["pdb_code"].lower()
        labels_heavy, labels_light = values["H_id labels 4.5"], values["L_id labels 4.5"]
        sequence_heavy, sequence_light = values["H_id sequence"], values["L_id sequence"]
        my_label_dict[pdb]["heavy labels"]="".join([str(each) for each in labels_heavy])
        my_label_dict[pdb]["light labels"]="".join([str(each) for each in labels_light])
        my_label_dict[pdb]["heavy sequence"]="".join([str(each) for each in sequence_heavy])
        my_label_dict[pdb]["light sequence"]="".join([str(each) for each in sequence_light])
        my_label_dict[pdb]=dict(my_label_dict[pdb])
my_label_dict=dict(my_label_dict)
```

```python
for pdb in paragraph_label_dict.keys():
    print(pdb)
    print(paragraph_label_dict[pdb]["heavy labels"])
    print(my_label_dict[pdb]["heavy labels"])
    print(paragraph_label_dict[pdb]["heavy sequence"])
    print(my_label_dict[pdb]["heavy sequence"])
    print("="*100)
```
